```python
# create a connection object 
import csv, sqlite3
con = sqlite3.connect("socioeconomic.db") # "socioeconomic.db" is name of the database in cloud

# create curser object to work with sql queries 
cur = con.cursor()
```

### Storing the dataset in a Table
 

```python

##### To analyze the data using SQL, it first needs to be stored in the database.

##### We will first read the csv files  from the given url  into pandas dataframes

##### Next we will be using the  df.to_sql() function to convert each csv file  to a table in sqlite  with the csv data loaded in it.

import pandas
df = pandas.read_csv('ChicagoCensusData.csv')
df.to_sql("chicago_socioeconomic_data", con, if_exists='replace', index=False,method="multi")

# similarly to crime and school tables
crime_df = pandas.read_csv('ChicagoCrimeData.csv')
crime_df.to_sql("crime", con, if_exists='replace', index=False,method="multi")

school_df = pandas.read_csv('ChicagoPublicSchools.csv')
school_df.to_sql("school", con, if_exists='replace', index=False,method="multi")
```

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'crime_df' with the crime data
# You can group the data by community area and count the number of crimes in each area
crime_counts = crime_df['COMMUNITY_AREA_NUMBER'].value_counts().sort_index()

# Define the x and y variables for the bar chart
community_areas = crime_counts.index
crime_count = crime_counts.values

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(community_areas, crime_count, color='skyblue')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Add labels and title
plt.xlabel('Community Area')
plt.ylabel('Total Crime Count')
plt.title('Total Crime Counts by Community Area')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have a DataFrame named 'crime_df' with the crime data
# Calculate the top N crime types
top_n = 5
top_crime_type_counts = crime_df['PRIMARY_TYPE'].value_counts().nlargest(top_n)
other_count = crime_df['PRIMARY_TYPE'].count() - top_crime_type_counts.sum()

# Extract the top N crime types and their counts
top_crime_types = top_crime_type_counts.index
top_crime_counts = top_crime_type_counts.values

# Append 'Other' as a separate category with its count
top_crime_types = top_crime_types.append(pd.Index(['Other']))
top_crime_counts = np.append(top_crime_counts, other_count)

# Create a pie chart
plt.figure(figsize=(10, 10))
plt.pie(top_crime_counts, labels=top_crime_types, autopct='%1.1f%%', startangle=200)

# Add a title
plt.title(f'Top {top_n} Crime Types and Other Distribution')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'crime_df' with the crime data
# You can group the data by year and count the number of crimes for each year
crime_counts_by_year = crime_df['YEAR'].value_counts().sort_index()

# Define the x and y variables for the line chart
years = crime_counts_by_year.index
crime_count = crime_counts_by_year.values

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(years, crime_count, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Total Crime Count')
plt.title('Crime Trends Over Time')

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'crime_df' with year and crime type columns
# Calculate the top N crime types
top_n = 5
top_crime_types = crime_df['PRIMARY_TYPE'].value_counts().nlargest(top_n).index

# Filter the DataFrame to include only the top N crime types and group the rest as 'Other'
filtered_crime_df = crime_df.copy()
filtered_crime_df.loc[~filtered_crime_df['PRIMARY_TYPE'].isin(top_crime_types), 'PRIMARY_TYPE'] = 'Other'

# Create a pivot table for crime types by year
crime_pivot = pd.crosstab(filtered_crime_df['YEAR'], filtered_crime_df['PRIMARY_TYPE'])

# Create a stacked bar chart
plt.figure(figsize=(12, 6))
crime_pivot.plot(kind='bar', stacked=True, colormap='viridis')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Count')
plt.title(f'Top {top_n} Crime Types and Other by Year (Stacked Bar Chart)')

# Show the plot
plt.legend(title='Crime Type')
plt.show()
```

```python
query = '''SELECT "Elementary, Middle, or High School" AS school_type, AVG("SAFETY_SCORE") AS average_safety_score
           FROM school
           GROUP BY "Elementary, Middle, or High School"
           ORDER BY average_safety_score DESC'''

r = pd.read_sql(query, con)
r
```

### Finding community area Name with the most number of crimes.

```python
query = """ select csd.community_area_name from chicago_socioeconomic_data as csd
where csd.community_area_number = (select community_area_number from crime group by community_area_number 
order by count(*) desc limit 1)"""
r  = pd.read_sql(query, con)
r
```

### Finding the community area with the highest hardship index and the most recorded crimes.

```python
query = """SELECT CSD.COMMUNITY_AREA_NAME, CSD.HARDSHIP_INDEX, COUNT(*) AS crime_count
FROM CHICAGO_SOCIOECONOMIC_DATA AS CSD
JOIN CRIME AS C ON CSD.COMMUNITY_AREA_NUMBER = C.COMMUNITY_AREA_NUMBER
GROUP BY CSD.COMMUNITY_AREA_NAME
ORDER BY CSD.HARDSHIP_INDEX DESC, crime_count DESC limit 5;
"""
r  = pd.read_sql(query, con)
r
```
